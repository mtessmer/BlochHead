```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')

# draw sphere
u, v = np.mgrid[0:2*np.pi:200j, 0:np.pi:100j]
xs = np.cos(u)*np.sin(v)
ys = np.sin(u)*np.sin(v)
zs = np.cos(v)
ax.set_box_aspect(aspect = (1, 1, 1))
plt.axis('off')
ax.plot([0, 0], [0, 0], [-1, 1], color='k')
ax.plot([0, 0], [-1, 1] , [0, 0], color='k')
ax.plot([-1, 1], [0, 0], [0, 0],  color='k')
ax.plot(xs[zs == np.abs(zs).min()], ys[zs == np.abs(zs).min()], np.zeros(200), color = 'k', alpha=0.2)
ax.plot(np.zeros(200), xs[zs == np.abs(zs).min()], ys[zs == np.abs(zs).min()], color = 'k', alpha=0.2)

vec = np.array([5., -5., 8.])
vec /= np.linalg.norm(vec)
print(vec)

ax.plot([0, vec[0]], [0, vec[1]], [0,vec[2]], color='tab:red', linewidth=4)
ax.plot([0, vec[0]], [vec[1], vec[1]], [0,0], color='tab:red', linestyle='--')
ax.plot([vec[0], vec[0]], [0, vec[1]], [0,0], color='tab:red', linestyle='--')
ax.plot([vec[0], vec[0]], [vec[1], vec[1]], [0, vec[2]], color='tab:red', linestyle='--')
ax.scatter(*vec, color='tab:red', s=100)

# ax.plot(ys[zs == np.abs(zs).min()], zs[zs == np.abs(zs).min()], xs[zs == np.abs(zs).min()], color = 'k')
# ax.plot(xs[xs == np.abs(xs).min()], ys[xs == np.abs(xs).min()], zs[xs == np.abs(xs).min()], color = 'k')


ax.plot_surface(xs, ys, zs, linewidth=1., rstride=1, cstride=1, cmap=plt.cm.viridis, alpha=0.2)
fig.savefig('bloch_sphere.svg', transparent=True)
```
